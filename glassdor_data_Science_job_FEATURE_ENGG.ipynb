## cleaning the data


```python
df['Salary Estimate'][:10]
```

```python
# in the salary estimates column we need to convert it to string and should remove the $ sign
df['Salary Estimate'][0].split('(')[0].replace('$',"").replace("K","")
```

```python
df['Salary Estimate'] = df['Salary Estimate'].apply(lambda x: x.split('(')[0].replace('$','').replace('K',''))
```

```python
df['Salary Estimate'].unique()
```

```python
# now we got the salary estimates. lets convert this column into minimum and maximum salary and also can create average salary column

df['Minimum Salary']=df['Salary Estimate'].apply(lambda x: x.split('-')[0]).astype(int)
df['Maximum Salary']=df['Salary Estimate'].apply(lambda x: x.split('-')[1]).astype(int)
df['Average Salary']=round((df['Minimum Salary']+df['Maximum Salary'])/2,2)
```

```python
df.head()
```

```python
df['Job Description'][0]
```

```python
# job description column contains lots of data.we need to separate it based on domain ex:data science , sql, machine learning

df['Python']=df['Job Description'].apply(lambda x: 1 if 'python' in x.lower() else 0)
df['SQL']=df['Job Description'].apply(lambda x: 1 if 'sql' in x.lower() else 0)
df['Machine Learning']=df['Job Description'].apply(lambda x: 1 if 'machine learning' or 'ml' in x.lower() else 0)
df['Hadoop']=df['Job Description'].apply(lambda x: 1 if 'hadoop' in x.lower() else 0)
df['Big Data']=df['Job Description'].apply(lambda x: 1 if 'big data' in x.lower() else 0)
df['Excel']=df['Job Description'].apply(lambda x : 1 if 'excel' in x.lower() else 0)
df['Tableau']=df['Job Description'].apply(lambda x: 1 if 'tableau' in x.lower() else 0)
df['AWS']=df['Job Description'].apply(lambda x: 1 if 'amazon web services' or 'aws' in x.lower() else 0)
df['Spark']=df['Job Description'].apply(lambda x:1 if 'spark' in x.lower() else 0)
```

```python
# converting job title column into job and seniority columns
def simp_job(title):
    if 'data scientist' in title.lower() or 'data' and 'scientist' in title.lower():
        return 'Data Scientist'
    elif 'data modeler' in title.lower():
        return 'Data Modeler'
    elif 'data analyst' in title.lower() or 'data' and 'analyst' in title.lower():
        return 'Data Analyst'
    elif 'data engineer' in title.lower() or 'data' and 'engineer' in title.lower():
        return 'Data Engineer'
    elif 'machine learning' in title.lower():
        return 'Machine Learning Engineer'
    elif 'director' in title.lower():
        return 'Director'
    elif 'manager' in title.lower():
        return 'Manager'
    elif 'analyst' in title.lower():
        return 'Analyst'
    else:
        return np.nan
```

```python
def seniority(tit):
    if 'sr' in tit.lower() or 'lead' in tit.lower() or 'senior' in tit.lower() or 'principal' in tit.lower():
        return 'Senior'
    elif 'jr' in tit.lower() or 'junior' in tit.lower():
        return 'Junior'
    elif 'associate' in tit.lower():
        return 'Associate'
    elif 'manager' in tit.lower():
        return 'Manager'
    else:
        return np.nan
```

```python
df['Job']=df['Job Title'].apply(simp_job)
df['Seniority']=df['Job Title'].apply(seniority)
```
